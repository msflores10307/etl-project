```python
diabetes_rate = "./Resources/diabetes_rate.csv"
diabetes_prev = "./Resources/diabetes_prev.csv"
rate = pd.read_csv(diabetes_rate)
prev = pd.read_csv(diabetes_prev, header=2)
rate
```

```python
prev_clean = prev[["Country Name", "2019"]].copy()
prev_clean = prev_clean.dropna()
prev_clean
```

```python
diabetes_df = pd.merge(prev_clean, rate, on="Country")
diabetes_df.dropna()
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://www.statista.com/statistics/1104709/coronavirus-deaths-worldwide-per-million-inhabitants/'
browser.visit(url)
time.sleep(5)
xpath = '/html/body/div[3]/div[4]/section[2]/div[2]/div[2]/div[2]/div[1]/div[1]/article/div/div/div[1]/div/div/div/div[2]/div[2]/button[2]'
cookies = '/html/body/div[3]/div[4]/div[48]/div/div/button'
master = pd.DataFrame([])
browser.find_by_xpath(cookies).first.click()
for x in range(13):
    html = browser.html
    datapage = pd.read_html(html)[0]
    master = pd.concat([master,datapage])
    browser.find_by_xpath(xpath).first.click()
```

```python
covid_stats = master.rename(columns={"Unnamed: 0": "Country", "Confirmed deaths (absolute)": "COVID Deaths"})
covid_stats = covid_stats[["Country", "COVID Deaths"]]
covid_stats
```
