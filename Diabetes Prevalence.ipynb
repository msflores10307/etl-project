```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
diabetes_rate = "./Resources/diabetes_rate.csv"
diabetes_prev = "./Resources/diabetes_prev.csv"
rate = pd.read_csv(diabetes_rate)
prev = pd.read_csv(diabetes_prev, header=2)
```

```python
prev_clean = prev[["Country Name", "2019"]].copy()
prev_clean = prev_clean.dropna()
prev_clean
```

```python
prev_clean = prev_clean.rename(columns={"Country Name": "Country", "2019": "Diabetes prevalence in 2019"})
rate = rate.rename(columns={"country": "Country", "pop2020": "Population"})
```

```python
diabetes_df = pd.merge(prev_clean, rate, on="Country")
diabetes_df.dropna()
```

```python
diabetes_df.to_csv("Resources/combined_data.csv")
```
